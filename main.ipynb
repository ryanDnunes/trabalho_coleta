CRIEI UM METODO PARA SABER OS TIPO DE SITUAÇÕES DE ELEIÇÃO

0       RENÚNCIA/FALECIMENTO/CASSAÇÃO APÓS A ELEIÇÃO
1                                        SUBSTITUÍDO
2                   REGISTRO NEGADO ANTES DA ELEIÇÃO
3                                           2º TURNO
4                                             #NULO#
5                                         NÃO ELEITO
6                             INDEFERIDO COM RECURSO
7                     REGISTRO NEGADO APÓS A ELEIÇÃO
8              RENÚNCIA/FALECIMENTO COM SUBSTITUIÇÃO
9                                             ELEITO
10    RENÚNCIA/FALECIMENTO/CASSAÇÃO ANTES DA ELEIÇÃO
dtype: object

```python
def criar_csv(sigla, ano):
    aux = f'resultado/votacao_candidato_munzona_{ano}/votacao_candidato_munzona_{ano}_{sigla}.csv'
    aux2 = f'candidato/consulta_cand_{ano}/consulta_cand_{ano}_{sigla}.csv'
    candidato = pd.read_csv(aux2, delimiter=';', encoding='latin1', low_memory=False)
    
    # Filtrar colunas e linhas de interesse
    colunas_vencedor = ['NM_CANDIDATO', 'NR_CANDIDATO', 'CD_CARGO', 'DS_SIT_TOT_TURNO']
    df_vencedor = candidato[colunas_vencedor]
    vencedor = df_vencedor[(df_vencedor['CD_CARGO'] == 11)].drop_duplicates()
    
    return vencedor['DS_SIT_TOT_TURNO']

# Lista de siglas e anos para iterar
lista_sigla = ['AC', 'AL', 'AM', 'AP', 'BA', 'BRASIL', 'CE', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']
anos_de_eleicao_para_prefeito = ['2000', '2004', '2008', '2012', '2016', '2020']

situacoes_unicas = set()

# Iterar sobre siglas e anos
for sigla in lista_sigla:
    for ano in anos_de_eleicao_para_prefeito:
        situacoes_unicas.update(criar_csv(sigla, ano).drop_duplicates())

# Exibir os valores únicos
print(pd.Series(list(situacoes_unicas)))
```

verifica se há valores nulos

```python
import pandas as pd

csv1 = 'dados_selecionas_por_mim/candidatos_vencedores.csv'
df = pd.read_csv(csv1, delimiter=';')  


def verificar_nulos(df):
    resultado = {}
    
    for coluna in df.columns:
        valores_nulos = df[coluna].isnull().sum()
        valores_hashtag_nulo = (df[coluna] == '#NULO').sum()
        
        # Adiciona ao resultado apenas se houver algum valor nulo ou '#NULO'
        if valores_nulos > 0 or valores_hashtag_nulo > 0:
            resultado[coluna] = {
                'valores_nulos': valores_nulos,
                '#NULO': valores_hashtag_nulo
            }
    
    return resultado

# Executa a função e imprime o resultado
resultado = verificar_nulos(df)
if resultado:
    print("Colunas com valores nulos ou '#NULO':")
    for coluna, contagens in resultado.items():
        print(f"Coluna: {coluna}, Valores nulos: {contagens['valores_nulos']}, '#NULO': {contagens['#NULO']}")
else:
    print("Não há valores nulos ou '#NULO' no arquivo.")
```
